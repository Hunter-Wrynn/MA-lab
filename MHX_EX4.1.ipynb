## 基于度连接

```python
import networkx as nx
import matplotlib.pyplot as plt
from IPython.display import clear_output

def degree_based_update(G):
    new_edges = []
    for node in G.nodes():
        neighbors = list(G.neighbors(node))
        non_neighbors = list(set(G.nodes()) - set(neighbors) - {node})
        if non_neighbors:
            max_degree_neighbor = max(non_neighbors, key=lambda x: G.degree(x))
            new_edges.append((node, max_degree_neighbor))
    return new_edges

def visualize_degree_based_update(G, pos, iterations):
    previous_edges = set()
    
    fig, axes = plt.subplots(iterations, 1, figsize=(6, iterations * 4))
    if iterations == 1:
        axes = [axes]
    
    for i in range(iterations):
        clear_output(wait=True)
        print(f"Updating and visualizing degree-based graph... Iteration {i+1}/{iterations}")

        new_edges = degree_based_update(G)
        G.remove_edges_from(list(G.edges()))
        G.add_edges_from(new_edges)

        nx.draw(
            G,
            pos,
            node_color="blue",
            edge_color="black",
            style="solid",
            node_shape="o",
            ax=axes[i]
        )
        nx.draw_networkx_edges(
            G, pos, edgelist=list(previous_edges), edge_color="green", style="solid", ax=axes[i]
        )
        axes[i].set_title(f"Iteration {i+1}")
        previous_edges = set(new_edges)

    plt.suptitle("Degree-Based Connection")
    plt.show()

# Example usage:
G = nx.gnp_random_graph(15, 0.3)
pos = nx.circular_layout(G)

visualize_degree_based_update(G, pos, iterations=10)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
from IPython.display import clear_output

def random_global_update(G):
    new_edges = []
    for node in G.nodes():
        neighbors = list(G.neighbors(node))
        non_neighbors = list(set(G.nodes()) - set(neighbors) - {node})
        if non_neighbors:
            random_neighbor = random.choice(non_neighbors)
            new_edges.append((node, random_neighbor))
    return new_edges

def visualize_random_global_update(G, pos, iterations):
    previous_edges = set()
    fig, axes = plt.subplots(iterations, 1, figsize=(6, iterations * 4))
    if iterations == 1:
        axes = [axes]
    
    for i in range(iterations):
        clear_output(wait=True)
        print(f"Updating and visualizing random global graph... Iteration {i+1}/{iterations}")

        new_edges = random_global_update(G)
        G.remove_edges_from(list(G.edges()))
        G.add_edges_from(new_edges)

        nx.draw(
            G,
            pos,
            node_color="blue",
            edge_color="black",
            style="solid",
            node_shape="o",
            ax=axes[i]
        )
        nx.draw_networkx_edges(
            G, pos, edgelist=list(previous_edges), edge_color="green", style="solid", ax=axes[i]
        )
        axes[i].set_title(f"Iteration {i+1}")
        previous_edges = set(new_edges)

    plt.suptitle("Random Global Connection")
    plt.show()


# Example usage:
G = nx.gnp_random_graph(15, 0.3)
pos = nx.circular_layout(G)

visualize_random_global_update(G, pos, iterations=10)
```
